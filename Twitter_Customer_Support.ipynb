```python
#For which company compalaints are maximum
import re
company = []
for ele in dt_cust_tweet.text:   
    try:
        company.append(re.search(r'^@[a-zA-Z0-9]+\s', ele).group())
    except:
        pass
company = [s.strip('@') for s in company]

keys, counts = np.unique(company, return_counts=True)
company_complaints = pd.DataFrame(list(zip(keys, counts)), 
               columns =['Company', '#Compalints'])
company_complaints.sort_values(by='#Compalints', ascending=False, inplace = True)
print(company_complaints.head(25))
company_complaints.describe()
```

```python
hastags = []
for k in dt_cust_tweet.text:   
    try:
        hastags.append(re.search(r'#[\w]*', k).group())
    except:
        hastags.append("")
hastags = [j.strip('#') for j in hastags]
dt_cust_tweet['hastag'] = hastags
```

```python
def clean_tweet(txt, pattern):
    tmp = re.findall(pattern, txt)
    for i in tmp:
        txt = re.sub(i, '', txt)
    return txt
dt_cust_tweet['cleaned_text'] = np.vectorize(clean_tweet)(dt_cust_tweet['text'], "@[\w]*") #Remove twitter account name
dt_cust_tweet['cleaned_text'] = dt_cust_tweet['cleaned_text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
dt_cust_tweet['cleaned_text'] = dt_cust_tweet['cleaned_text'].str.replace("[^a-zA-Z]", " ") #Taking only characters
 
dt_cust_tweet.head(20)
```

```python
len(dt_cust_tweet['hastag'].unique())
```

```python
dt_cust = dt_cust_tweet[["cleaned_text"]]
dt_cust["cleaned_text"] = dt_cust["cleaned_text"].astype(str)
dt_cust["cleaned_text"] = dt_cust["cleaned_text"].str.lower()
dt_cust.head(10)
```

```python
#Remove Stopwords

#import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

dt_cust["cleaned_text"] = dt_cust["cleaned_text"].apply(lambda text: remove_stopwords(text))
dt_cust.head(10)
```

```python
#Removing small words (<4)
dt_cust["cleaned_text"] = dt_cust["cleaned_text"].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

#Text normalization
from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_tweet = dt_cust["cleaned_text"].apply(lambda x: x.split()) # tokenizing

tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_tweet.head(10)
```

```python
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
    
dt_cust["cleaned_tokenize_text"] = tokenized_tweet
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from watson_developer_cloud import NaturalLanguageUnderstandingV1
from watson_developer_cloud.natural_language_understanding_v1 import Features, EntitiesOptions, KeywordsOptions, SentimentOptions, CategoriesOptions
```

```python
#Vadar Sentiment Analysis
analyser = SentimentIntensityAnalyzer()

#for IBM sentiment analysis
natural_language_understanding = NaturalLanguageUnderstandingV1(                                         
    version='2018-11-16',
    iam_apikey='deleted', 
    url='deleted' 
    )
```

```python
dt_cust['cleaned_text'].replace('', np.nan, inplace=True)
dt_cust.dropna(inplace = True)
dt_cust.head()
```

```python
def analyze_text(input_text, analyzer):
    if analyzer == 'VADER':
        result = analyser.polarity_scores(input_text)
        score = result['compound']
    elif analyzer == 'TextBlob':
        score = TextBlob(input_text).sentiment.polarity
    else:
        pass
#        response = natural_language_understanding.analyze(text = input_text , features = 
#                                                         Features(sentiment = SentimentOptions()), language='en').get_result()
#        score = response.get('sentiment').get('document').get('score')
    
    if score > 0:
        result = 'Possitive'
    else:
        result = 'Negative'
    
    return result

#Creating new column
dt_cust['Vadar_sent'] = dt_cust['cleaned_text'].apply(analyze_text, analyzer = 'VADER')
dt_cust['TextBlob_sent'] = dt_cust['cleaned_text'].apply(analyze_text, analyzer = 'TextBlob')
#dt_cust['IBM_sent'] = dt_cust['cleaned_text'].apply(analyze_text, analyzer = 'IBM')
dt_cust['Vadar_tokenize_sent'] = dt_cust['cleaned_tokenize_text'].apply(analyze_text, analyzer = 'VADER')
dt_cust['TextBlob_tokenize_sent'] = dt_cust['cleaned_tokenize_text'].apply(analyze_text, analyzer = 'TextBlob')
#dt_cust['IBM_tokenize_sent'] = dt_cust['cleaned_tokenize_text'].apply(analyze_text, analyzer = 'IBM')

dt_cust.head(10)
```
